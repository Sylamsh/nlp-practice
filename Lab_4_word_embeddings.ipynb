**Task 4:** Write a module to plot the word vectors in a 2-D vector space


*   You need to reduce the 300 vector size (dimension) to a 2-D vector using [TSNE](https://scikit-learn.org/stable/modules/generated/sklearn.manifold.TSNE.html?highlight=tsne#sklearn.manifold.TSNE).
*   Each word is now represented by this approximated 2-D vector which you can consider as a point of 2 coordinates, where the first vector element is the x-coordinate and the second vector element is the y-coordinate. Finally, you can plot a scatter plot using your prefferred plotting libary. 



````python
#####################################
############ Code cell #7 ###########
#####################################
# we a list of words and extract their corresponding vectors from the word2vec model 
vocab = ["boy", "girl", "man", "woman", "king", "queen", "banana", "apple", "mango", "fruit", "coconut", "orange"]

# the function takes the model as the argument and plots the scatter plot graph
def tsne_plot(model):
    labels = [] # list to append the labels, labels is the word in the vocab list
    wordvecs = [] # list to append the extracted word vectors for the respective labels/words

    ## complete this loop
    for word in vocab: # iterate over the vocabs
        ## Todo #1 : append each word vectors of the current word into the wordvecs list
        wordvecs.append(word2vec_model[word])
        ## Todo #2 : append each labels or the current word into the labels list
        labels.append(word)
        ## HinT: you can access the word vectors using ``word2vec_model[word]`` as done in ``Code cell #3``

    # Reducing the word vectors from 300 to 2 dimensioality for visualizing in a 2-D space using TSNE
    tsne_model = TSNE(perplexity=3, n_components=2, init='pca', random_state=42)
    # complete the following to fit and transform the word vectors and store into the coordinates variable
    coordinates = tsne_model.fit_transform(wordvecs)
    x = coordinates[:, 0] # initialise an empty list to store the x-coordinate from the coordinates variable
    y = coordinates[:, 1] # initialise an empty list to store the y-coordinate from the coordinates variable
    
    # complete the following loop to append each x and y values of the coordinates so that they can be used for the scatter plot.

    # for value in coordinates:
    #     x.append(value[0]) ## the first element of the coordinate value should go into the list of x-coordinate values
    #     y.append(value[1]) ## the second element of the coordinate value should go into the list of y-coordinate values
        
    plt.figure(figsize=(8,8)) 
    for i in range(len(x)):
        plt.scatter(x[i],y[i])
        plt.annotate(labels[i],
                     xy=(x[i], y[i]),
                     xytext=(2, 2),
                     textcoords='offset points',
                     )
    plt.show()

tsne_plot(word2vec_model)
````

## Training your own word2vec model

````python
# using 20newsgroups dataset from sklearn

from sklearn.datasets import fetch_20newsgroups

dataset = fetch_20newsgroups(shuffle=True, random_state=1, remove=('headers', 'footers', 'quotes'))
documents = dataset.data
````

````python
# tokenize the documents
from nltk.tokenize import sent_tokenize, word_tokenize
import nltk
nltk.download('punkt')

token_list = []
for d in documents:
    s = sent_tokenize(d) 
    token_list = token_list +[word_tokenize(t) for t in s]

# check the first three sentences
token_list[:3]
````

````python
# Training the custom word2vec model
from gensim.models.word2vec import Word2Vec

# Number of vector elements (dimensions) to represent the word vector
num_features = 50

model = Word2Vec(token_list, size=num_features)
````

````python
# dicarding unneccesary weights
model.init_sims(replace=True)
````

````python
# saving the trained model for later use
model_name = "custom_model"
model.save(model_name)
````

````python
movie_data = load_files("txt_sentoken")
````

````python
X, y = movie_data.data, movie_data.target
print(f'Total number of documents {len(X)}')
````

````python
## Preprocessing the data
documents = []
from nltk.stem import WordNetLemmatizer

stemmer = WordNetLemmatizer()

for sen in range(0, len(X)):
    ## Remove all the special characters or the non-word characters
    document = re.sub(r'\W', ' ', str(X[sen]))
    
    ## remove all single characters
    document = re.sub(r'\s+[a-zA-Z]\s+', ' ', document)
    ## Substituting multiple spaces with single space
    document = ' '.join(document.split())
        
    ## Converting to Lowercase
    document = document.lower()
    documents.append(document)

# tokenize the documents
token_list = []
for d in documents:
    s = sent_tokenize(d) 
    token_list = token_list +[word_tokenize(t) for t in s]

# check the first three sentences
token_list[:3]
````

````python
# Training the custom word2vec model
num_features = 300
min_word_count = 100 
num_workers = 4
window = 10
downsampling = 1e-3
model = Word2Vec(token_list, size=num_features)
````

````python
# dicarding unneccesary weights
model.init_sims(replace=True)
````

````python
# saving the trained model for later use
model_name = "movie_review_w2v"
model.save(model_name)
````

## Utility functions for cleanup

````python
from bs4 import BeautifulSoup

def clean_sentence(sentence, remove_stopwords):
    removed_markup = BeautifulSoup(sentence, 'html.parser').get_text()
    removed_punctuation = re.sub(r'[^a-zA-Z0-9]', ' ', removed_markup)
    tokens = removed_punctuation.lower().split()
    if remove_stopwords:
        tokens = [t for t in tokens if t not in stopwords]
    return tokens
````

````python
tokenizer = nltk.data.load('tokenizers/punkt/english.pickle')

def review_to_sentences(review, remove_stopwords):
    raw_sentences = tokenizer.tokenize(review.strip())
    return [clean_sentence(sentence, remove_stopwords) for sentence in raw_sentences if len(sentence) > 0]
````

````python
def compute_sentence_vec(sentence):
    word_vecs = [model.wv[word] for word in sentence if word in model.wv.vocab]
    
    if len(word_vecs) == 0:
        return np.zeros([model.vector_size])
    
    feature_vec = np.stack(word_vecs)
    return np.average(feature_vec, axis=0)

def compute_review_vec(review):
    feature_vec = np.stack([compute_sentence_vec(sentence) 
                            for sentence in review_to_sentences(review, remove_stopwords=True)
                            if len(sentence) > 0])
    return np.average(feature_vec, axis=0)
````

### Obtaining Word2Vec

````python
vecs = np.stack([compute_review_vec(review) for review in documents])
vecs.shape
````

# Classifiers

````python
# train-test split
X_train, X_test, y_train, y_test = train_test_split(vecs, y, test_size=0.2, random_state=0)
````

````python
# utility function for evaluating classifier performance
def evaluate_performance(clf, clf_name):
  clf.fit(X_train, y_train)
  y_pred = clf.predict(X_test)
  print(f"Accuracy | Word2Vec | {clf_name}: {str(accuracy_score(y_test, y_pred)*100)} %")
````
